```python
import requests
from bs4 import BeautifulSoup

def scrape_cnbc_news():
    url = "https://www.cnbc.com/world/?region=world"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    articles = soup.find_all('div', class_='LatestNews-headlineWrapper')
    
    news_data = []
    for article in articles:
        heading = article.find('a').text
        time_element = article.find('time')
        date = time_element['time'] if time_element and 'datetime' in time_element.attrs else 'No date provided'
        news_link = article.find('a')['href']
        news_data.append({'Heading': heading, 'Date': date, 'News Link': news_link})
    
    df = pd.DataFrame(news_data)
    return df

cnbc_df = scrape_cnbc_news()
print(cnbc_df)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_bewakoof():
    url = "https://www.bewakoof.com/bestseller?sort=popular"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    products = soup.find_all('div', class_='productCardBox', limit=10)
    
    product_data = []
    for product in products:
        name = product.find('div', class_='productCardDetail').h3.text
        price_element = product.find('span', class_='discountedPriceText')
        price = price_element.text if price_element else 'Price not available'
        image_url = product.find('img', class_='productImgTag')['src']
        product_data.append({'Product Name': name, 'Price': price, 'Image URL': image_url})
    
    df = pd.DataFrame(product_data)
    return df

bewakoof_df = scrape_bewakoof()
print(bewakoof_df)
```
